Revisit the programming task for "**Activity 4 Exercise 1**" where you emulated a stack and queue using a list.  Consider how these programs could be combined and how functions could be used to make the code more concise and reusable.
````python
# Adapting 4.1 to work as functions
queue = [1,2,3,4,5]
stack = [1,2,3,4,5]

def addToQueue():
    item = input("Input item to add: ")
    queue.append(item)

def removeFromQueue():
    if len(queue) != 0:
        print(f"Removed Item: {queue[0]}")
        queue.remove(queue[0])
    else:
        print("Queue is empty")

def printQueue():
    if len(queue) != 0:
        for i in queue:
            print(i, end=" ")
        print("\n")
    else:
        print("Queue is empty")
        
def queueMenu():
    inQueue = True
    while inQueue:
        print(("Please select from the following:): "))
        print("\t1. Add to queue")
        print("\t2. Remove from queue")
        print("\t3. View queue")
        print("\t8. Return to main")
    
        userInput = int(input(">> "))
    
        if userInput == 1:
            addToQueue()
        elif userInput == 2:
            removeFromQueue()
        elif userInput == 3:
            printQueue()
        elif userInput == 8:
            inQueue = False
        else:
            print("Incorrect selection, please try again")

def pushStack():
    item = input("Input item to add: ")
    stack.append(item)

def popStack():
    if len(stack) != 0:
        print(f"Next Item: {stack.pop()}")
    else:
        print("Stack is empty")

def printStack():
    if len(stack) != 0:
        for i in stack:
            print(i, end=" ")
        print("\n")
    else:
        print("Stack is empty")
        
def stackMenu():
    inStack = True
    while inStack:
        print(("Please select from the following:): "))
        print("\t1. Push onto stack")
        print("\t2. Pop stack")
        print("\t3. View stack")
        print("\t8. Return to main")
        
        userInput = int(input(">> "))
        
        if userInput == 1:
            pushStack()
        elif userInput == 2:
            popStack()
        elif userInput == 3:
            printStack()
        elif userInput == 8:
            inStack = False
        else:
            print("Incorrect selection, please try again")   
            
progContinue = True 
while progContinue:
    print(("Please select from the following:): "))
    print("\t1. Manage queue")
    print("\t2. Manage stack")
    print("\t9. Exit")
    
    userInput = int(input(">> "))
    
    if userInput == 1:
        queueMenu()
    elif userInput == 2:
        stackMenu()
    elif userInput == 9:
        progContinue = False
    else:
        print("Incorrect selection, please try again")
        
print("Exiting program.")
````

````python
class Name(object):
    def __init__(self, title, fName, sName, *oNames):
        self.firstName = fName
        self.surname = sName
        self.title = title # this could be defined as an enumerated type
        
        # what do you think the * operator is doing?
        if len(oNames) != 0:
            self.otherName = oNames
        else:
            self.otherName = [] # so stuff can be added later
    
    #Mr J. S. Greenhold Mr J. Greenhold
    # assuming all attributes hold valid data except other names
    def formalName(self):
        name = self.title + " " +self.firstName.upper()[0] + "."
        
        if len(self.otherName) != 0:
            for n in self.otherName:
                name = name + " " + n.upper()[0] + "."
        name = name + " " + self.surname.title()
        return name

lorenzoName = Name("Mr","Lorenzo","Dubois","Ruelle", "Garlen")
takeshiName = Name("Mr","Takeshi","Tanaka")
print(lorenzoName.formalName())
print(takeshiName.formalName())
````
