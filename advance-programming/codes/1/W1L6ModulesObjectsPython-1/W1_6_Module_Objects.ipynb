```python
# This is skecth (incomplete) of a class
# representing a data of birth
# You will need to run this cell BEFORE you run the next cell
# Does not do anything when run
class DOB(object):
    def __init__(self, dobString):
        # the default delimeter
        self.delim = '/'

        # split the string and make values int
        date = dobString.split(self.delim) 
        chDay = int(date[0])
        chMonth = int(date[1])
        chYear = int(date[2])
    
        # checking variables for determining
        # if to create using passed value OR
        # create a default (empty) DOB object
        ok = False
        length = False 

        # checking the validit of the data passed in
        if len(date) <= 3:
            length = True
            ok = DOB.checkDate(chDay, chMonth , chYear)

        # values are internaly stored as ints 
        if ok and length:
            self.day = chDay
            self.month = chMonth
            self.year = chYear
        else:
            self.day = -1
            self.month = -1
            self.year  = -1
      
      # used internaly to check any values changed  
    def checkDate(chDay, chMonth, chYear):
        valid = False
        # assumes that older is not valid
        if chYear >= 1900:
            
            # Test valid number of days given the month
            if(chMonth >=1 and chMonth<= 12) and (chDay >= 1 and chDay <=31):
               
                # 30 days has Sept 9, April 4, June 6, Nov 11
                if(chMonth == 4 or chMonth == 6 or chMonth == 9 
                or chMonth == 11) and chDay <= 30:
                    valid = True
                
                # all the rest have 31
                elif (chMonth == 1 or chMonth == 3 or chMonth == 5 or chMonth == 7 
                or chMonth == 8 or chMonth == 10 or chMonth == 12) and chDay <= 31 :
                    valid = True
                # except February 2
                elif chMonth == 2 and chDay <= 29:
                # 29 days in leap
                    if chYear%4 == 0 and chDay == 29:
                        valid = True
                    # feb with 28 days
                    elif chDay <= 28:
                        valid = True
        return valid 
      
    # returns an int    
    def getDay(self):
        return self.day

    # attempts to convert imput to an int
    # before checking
    def setDay(self, chDay):
        if type(chDay) == 'str':
            chDay = int(chDay)
            ok = DOB.checkDate(chDay, self.month, self.year)
            
            if ok:
                self.day = chDay


    # returns an int   
    def getMonth(self):
        return self.month
  
  # attempts to convert imput to an int
  # before checking
    def setMonth(self, chMonth):
        if type(chMonth) == 'str':
            chDay = int(chMonth)
            ok = DOB.checkDate(self.day, chMonth, self.year)
            
            if ok:
                self.month = chMonth
  
  
    # returns an int    
    def getYear(self):
        return self.year

    
    # attempts to convert imput to an int
    # before checking  
    def setYear(self, chYear):
        if type(chYear) == 'str':
            chDay = int(chYear)
            ok = DOB.checkDate(self.day, self.month, chYear)
            if ok:
                self.year = chYear

    # returns true if any DOB attributes are currently set to -1
    def isDefault(self):
        if  self.day == -1 or self.month == -1 or self.year == -1:
            return True
        else:
            return False 
  
    # constructs a string for output given current attributes
    def toString(self):
        return str(self.day) + self.delim + str(self.month) + self.delim + str(self.year)
```
