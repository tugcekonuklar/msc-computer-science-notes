### Exercise one

Using the following PeopleTrainingData.csv file write a Python program to parse (split) the content using string functions only and store it in an appropriate data structure. Print out a table with all the data using the following column headings (which are the first line in the css file): Title,Name,ID,Email,Company,Updated. Remember to consider readability and use formatting instructions in the output. 

```python
# Please note that the run order in this program is important.
# Things that change the order of rows or columns will break
# the whole program if run more than once.

# first two functions used to find column widths
# this could be done more consisely 
def colSizeComp(last, new):
    if last >= new:
        return last
    else:
        return new

    
def findColSize(data):
    # find how many columns
    cols = len(data[0]) 
    widths = [0] * cols
    #print(widths)
    for row in data:
        for i in range(0,cols):
            #print (i, "current:",widths[i], "next:",len(row[i]))
            widths[i] = colSizeComp(widths[i], len(row[i]))
    return widths


# print the data set within a table structure
def printDataSet(data):
    # to print the data set acuratly need to know max column width
    colW = findColSize(data)
    
    for row in data:
    # added extra space to all columns
        for i in range(0,len(row)):
            # justification here changed so name and surname sit together
            if i != 1 and i != 2:
                print(f"{row[i]:<{colW[i]+1}}", end='')
            if i == 1:
                print(f"{row[i]:>{colW[i]}}{row[i+1]:<{colW[i+1]}}", end='')
            # nothing is done when i = 2
        print()
             

# could be seperated into import, split, and clean as seperate functions
def cleanData(file):
    tempData = []
    lineCount = 0
    for line in file:
        # removes trailing white space
        if line != ' ' and line != '\n' and line != '':
            lineCount +=1
            # removes trailing white space
            line = line[0:len(line)-1]
            row = line.split(',')

        # column 1 & 2 has an extra character that
        # that needs to be removed but not on the first row
        if lineCount != 1:
            row[1] = row[1][1:len(row[1])-1]
            row[2] = row[2][0:len(row[2])-2]
        # first line is missing a heading
        if lineCount == 1:
            row.insert(2, ' ')

        #print(row)
        tempData.append(row)
    return tempData
        
file = open("PeopleTrainngDate.csv")
dataSet = cleanData(file)
printDataSet(dataSet)
       
```

```python
# this program requires the previous one to be run.
# only for import, split and clean.
# opted to handel the date problem first and then mo
from datetime import date

def upgradeDate(data, header):
    # upgrading the dataset so it contians a date object
    # instead of a string in the last column
    # you may have chosen to write a method to 
    # sort the data in its orginal structure
    for row in data:
        # skip first row if it contains header
        if header:
            header = False
            continue
        # first column now conatins the date
        splitS = row[0].split('/')
        #print(splitS)

        # I am assuming a UK date system in the file
        # which is day, month, year
        # date object = year, month, day
        newDate = date(int(splitS[2]), int(splitS[1]), int(splitS[0]))
        #print(str(newDate))
        row[0] = newDate
        print(row)
    return data

dataSet = upgradeDate(dataSet, True)

#Testing
for row in dataSet:
    print(row)

```

```python
# soritng the data
# the file provied is currently already sorted by data 
# you may want to manualy mix it up to demonstarte this works
import datetime

# This can be done more effectively with a lambda expresion (see week3)
def sort(data, rowStart, col): 
    l = len(data) 
    for i in range(rowStart, l): 
        for j in range(rowStart, l-1): 
            if (data[j][col] > data[j + 1][col]): 
                temp = data[j] 
                data[j]= data[j + 1] 
                data[j + 1]= temp 
    return data 

dataSet = sort(dataSet, 1, 0)

for row in dataSet:
    print(row)
 
```

```python
# outputting to file, I have done as a CSV file to demo this, however this was not specified

file = open("RestructuredPTD.csv", 'w')

def outputToFile(data, file, header):    
    colCount = len(data[0])
    # print header first
    if header:
        for i in range(0, colCount):
            if(i == colCount-1):
                file.write(data[0][i])
                file.write('\n')
            else:
                file.write(data[0][i])
                file.write(',')
    for row in data:
        count = 0
        # skip the first row
        if header:
            header = False
            continue
        else:
            for col in row:
                # last item in row need new line
                if count == colCount-1:
                    file.write(col)
                    file.write('\n')
                elif count == 0:
                    file.write(str(col))
                    file.write(',')
                    count += 1
                else:
                    file.write(col)
                    file.write(',')
                    count += 1
        
                
outputToFile(dataSet, file, True)
```
